**Исходные данные**
Для анализа использованы два файла Excel с вакансиями:
- `vacancies_ds.xlsx` – данные о вакансиях для Аналитиков данных.
- `vacancies_da.xlsx` – данные о вакансиях для Бизнес-аналитиков.

Оба файла содержат следующие поля:
- `id`: Уникальный идентификатор вакансии.

- `published_at`: Дата публикации.

- `department`: Отдел работодателя.
- `area`: Регион места работы.
- `experience`: Требуемый опыт работы.
- `key_skills`: Ключевые навыки.
- `schedule`: График работы.
- `employment`: Тип занятости.
- `description`: Описание вакансии.
- `description_lemmatized`: Лемматизированное описание вакансии.
- `salary_from`: Нижняя граница предлагаемой заработной платы.
- `salary_to`: Верхняя граница предлагаемой заработной платы.
- `salary_bin`: Категория зарплаты.
- `key_skills_from_key_skills_field`: Ключевые навыки из поля вакансии.
- `hard_skills_from_description`: Твердые навыки.
- `soft_skills_from_description`: Мягкие навыки.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds_file_path = '/content/drive/MyDrive/Colab Notebooks/vacancies_da.xlsx'
da_file_path = '/content/drive/MyDrive/Colab Notebooks/vacancies_ds.xlsx'

vacancies_ds = pd.read_excel(ds_file_path)
vacancies_da = pd.read_excel(da_file_path)

vacancies_ds.info(), vacancies_da.info()
```

```python
# Проверка и предобработка данных

# Проверим данные на наличие пропущенных значений и дубликатов
missing_values_ds = vacancies_ds.isnull().sum()
missing_values_da = vacancies_da.isnull().sum()
duplicates_ds = vacancies_ds.duplicated().sum()
duplicates_da = vacancies_da.duplicated().sum()

print('Таблица DS, пропуски:', missing_values_ds)
print('Таблица DA, пропуски:', missing_values_da)
print('Дубликаты в таблице DS:', duplicates_ds)
print('Дубликаты в таблице DA:', duplicates_da)

# Удалим дубликаты, если они есть
vacancies_ds_cleaned = vacancies_ds.drop_duplicates()
vacancies_da_cleaned = vacancies_da.drop_duplicates()

# Заполним пропущенные значения
# Заменим пропущенные значения в salary на медианные значения
vacancies_ds_cleaned['salary_from'].fillna(vacancies_ds_cleaned['salary_from'].median(), inplace=True)
vacancies_ds_cleaned['salary_to'].fillna(vacancies_ds_cleaned['salary_to'].median(), inplace=True)

vacancies_da_cleaned['salary_from'].fillna(vacancies_da_cleaned['salary_from'].median(), inplace=True)
vacancies_da_cleaned['salary_to'].fillna(vacancies_da_cleaned['salary_to'].median(), inplace=True)

# Проверим, что пропущенные значения заполнены
missing_values_ds_cleaned = vacancies_ds_cleaned.isnull().sum()
missing_values_da_cleaned = vacancies_da_cleaned.isnull().sum()

print()
print('Таблица DS, пропуски после очистки:', missing_values_ds_cleaned)
print('Таблица DA, пропуски после очистки:', missing_values_da_cleaned)
```

**Вывод**: Дубликатов в данных не обнаружили. Нашлись пропуски в категориальных полях: department, `key_skills_from_key_skills_field`, `hard_skills_from_description`, `soft_skills_from_description`. Причина в методе заполнения этих полей работадателями или HR-менеджеров.

Пропуски в количественных полях `salary_from` и `salary_to` заменили на медианные значения, чтобы не повлияли выбросы.

1. Анализ распределения грейдов (Junior, Junior+, Middle, Senior) по вакансиям↓

```python
# Анализ ключевых навыков для различных грейдов. Посмотрим распределение по кол-ву грейдов

plt.figure(figsize=(7, 4)),
plt.hist(vacancies_ds_cleaned['experience'].dropna(),bins= None)
plt.xlabel('Experience')
plt.ylabel('Count')
plt.title('Распределение опыта работы для DS')
plt.show()


plt.figure(figsize=(7, 4)),
plt.hist(vacancies_da_cleaned['experience'].dropna(),bins= None)
plt.xlabel('Experience')
plt.ylabel('Count')
plt.title('Распределение опыта работы для DA')
plt.show()

#считаем доли от общего числа
total_ds = vacancies_ds_cleaned['experience'].value_counts()
total_da = vacancies_da_cleaned['experience'].value_counts()

share_ds = 100 * total_ds / total_ds.sum()
share_da = 100 * total_da / total_da.sum()

print(share_ds)
print(share_da)
```

Вывод: В обоих профессиях очень низкая доля вакансий уровня Junior без опыта. Это значит, что компании не стремятся брать неопытных стажёров на эти позиции. 44% в DS и 61% составляют jun+ позиции. Это показывает, что компании берут потенциальных кандидатов с опытом работы от 1 года и вероятность устроится будет выше.

2. Анализ ключевых навыков для различных грейдов↓


```python
import ast

# Преобразование строкового представления списка в настоящий список для DA
vacancies_da_cleaned['key_skills'] = vacancies_da_cleaned['key_skills'].apply(
    lambda x: [skill.strip() for skill in ast.literal_eval(x)] if isinstance(x, str) else []
)

# Преобразование строкового представления списка в настоящий список для DS
vacancies_ds_cleaned['key_skills'] = vacancies_ds_cleaned['key_skills'].apply(
    lambda x: [skill.strip() for skill in ast.literal_eval(x)] if isinstance(x, str) else []
)

# Разделение навыков на отдельные строки для DA
key_skills_da_parsed = vacancies_da_cleaned.explode('key_skills')

# Разделение навыков на отдельные строки для DS
key_skills_ds_parsed = vacancies_ds_cleaned.explode('key_skills')

# Создание сводных таблиц по полю experience для DA
key_skills_pivot_da_parsed = key_skills_da_parsed.pivot_table(
    index='key_skills', columns='experience', aggfunc='size', fill_value=0
)

# Создание сводных таблиц по полю experience для DS
key_skills_pivot_ds_parsed = key_skills_ds_parsed.pivot_table(
    index='key_skills', columns='experience', aggfunc='size', fill_value=0
)

# Анализ топ-10 ключевых навыков для каждого уровня профессии DA
top_skills_da_parsed = {}
for exp in key_skills_pivot_da_parsed.columns:
    top_skills_da_parsed[exp] = key_skills_pivot_da_parsed[exp].sort_values(ascending=False).head(10)

# Анализ топ-10 ключевых навыков для каждого уровня профессии DS
top_skills_ds_parsed = {}
for exp in key_skills_pivot_ds_parsed.columns:
    top_skills_ds_parsed[exp] = key_skills_pivot_ds_parsed[exp].sort_values(ascending=False).head(10)

# Вывод результатов для DA
for exp, skills in top_skills_da_parsed.items():
    print(f'Top 10 Key Skills for {exp} (Data Analysis):')
    print(skills)
    print("\n")

# Вывод результатов для DS
for exp, skills in top_skills_ds_parsed.items():
    print(f'Top 10 Key Skills for {exp} (Data Science):')
    print(skills)
    print("\n")
```

### Определение, какие навыки спрашивают чаще - твердые или мягкие. К какому грейду и к какой специальности требований больше.

```python
# Обработка данных в столбцах hard_skills_from_description и soft_skills_from_description
vacancies_da_cleaned['hard_skills'] = vacancies_da_cleaned['hard_skills_from_description'].apply(
    lambda x: x.split(', ') if isinstance(x, str) else []
)
vacancies_da_cleaned['soft_skills'] = vacancies_da_cleaned['soft_skills_from_description'].apply(
    lambda x: x.split(', ') if isinstance(x, str) else []
)

vacancies_ds_cleaned['hard_skills'] = vacancies_ds_cleaned['hard_skills_from_description'].apply(
    lambda x: x.split(', ') if isinstance(x, str) else []
)
vacancies_ds_cleaned['soft_skills'] = vacancies_ds_cleaned['soft_skills_from_description'].apply(
    lambda x: x.split(', ') if isinstance(x, str) else []
)

# Разделение навыков на отдельные строки для DA и DS
hard_skills_da_parsed = vacancies_da_cleaned.explode('hard_skills')
soft_skills_da_parsed = vacancies_da_cleaned.explode('soft_skills')

hard_skills_ds_parsed = vacancies_ds_cleaned.explode('hard_skills')
soft_skills_ds_parsed = vacancies_ds_cleaned.explode('soft_skills')

# Функция для анализа навыков
def analyze_skills(hard_data, soft_data, title):
    hard_counts_by_experience = hard_data.groupby(['experience']).size()
    soft_counts_by_experience = soft_data.groupby(['experience']).size()

    print(f'\nНавыки для {title}:')
    print('Твердые навыки:')
    print(hard_counts_by_experience)
    print('Мягкие навыки:')
    print(soft_counts_by_experience)

    combined_counts = pd.DataFrame({
        'Твердые навыки': hard_counts_by_experience,
        'Мягкие навыки': soft_counts_by_experience
    }).fillna(0)

    combined_counts.plot(kind='bar', figsize=(5, 3), title=f'Навыки для {title}', stacked=False)
    plt.xlabel('Опыт работы')
    plt.ylabel('Количество навыков')
    plt.show()

# Анализ для Data Analysis
analyze_skills(hard_skills_da_parsed, soft_skills_da_parsed, 'Data Analysis')

# Анализ для Data Science
analyze_skills(hard_skills_ds_parsed, soft_skills_ds_parsed, 'Data Science')
```

Вывод.
Частота упоминания навыков:

Для обеих специальностей чаще всего упоминаются навыки у кандидатов с опытом Junior+ (1-3 года) и Middle (3-6 лет).
У Data Analysis наблюдается небольшой перевес в сторону мягких навыков для кандидатов с опытом Middle (3-6 лет), в то время как для Senior (6+ лет) мягкие навыки также превалируют, но разница менее значительная.
Для Data Science твердые навыки более востребованы для Junior+ (1-3 года) и Middle (3-6 лет), однако на уровне Senior (6+ лет) мягкие навыки упоминаются чаще.
Требования по грейдам:

Junior (no experience): На этом уровне для Data Science требования к твердым и мягким навыкам почти равны, однако для Data Analysis требования к мягким навыкам слегка превалируют.
Junior+ (1-3 года опыта): Для обеих специальностей как твердые, так и мягкие навыки имеют высокую частоту упоминания, с небольшим перевесом в сторону мягких навыков.
Middle (3-6 лет опыта): Здесь для Data Analysis мягкие навыки более востребованы, в то время как для Data Science требования к навыкам примерно равны.
Senior (6+ лет опыта): На этом уровне мягкие навыки упоминаются чаще для обеих специальностей, особенно для Data Science.
В целом, мягкие навыки имеют тенденцию становиться более важными на высших уровнях опыта, особенно для Data Science. Для Data Analysis ситуация более сбалансирована между твердыми и мягкими навыками на всех уровнях опыта, с небольшим перевесом в сторону мягких навыков.

### Определение наиболее желаемых кандидатов на вакансии Аналитик данных и Бизнес-аналитик по следующим параметрам: самые важные hard-skils, самые важные soft-skils.

```python
# Преобразование данных в списки и их разделение
vacancies_da_cleaned['key_skills'] = vacancies_da_cleaned['key_skills_from_key_skills_field'].dropna().str.split(',')
vacancies_ds_cleaned['key_skills'] = vacancies_ds_cleaned['key_skills_from_key_skills_field'].dropna().str.split(',')

# Разделение навыков на отдельные строки
key_skills_da_parsed = vacancies_da_cleaned.explode('key_skills')
key_skills_ds_parsed = vacancies_ds_cleaned.explode('key_skills')

# Очистка лишних пробелов
key_skills_da_parsed['key_skills'] = key_skills_da_parsed['key_skills'].str.strip()
key_skills_ds_parsed['key_skills'] = key_skills_ds_parsed['key_skills'].str.strip()

# Создание сводных таблиц для подсчета навыков по уровню опыта
key_skills_da_pivot = pd.pivot_table(key_skills_da_parsed, index=['experience', 'key_skills'], values='id', aggfunc='count').reset_index()
key_skills_ds_pivot = pd.pivot_table(key_skills_ds_parsed, index=['experience', 'key_skills'], values='id', aggfunc='count').reset_index()

# Функция для визуализации топ-10 навыков по уровню опыта
def visualize_top_skills_by_experience(skills_pivot, title):
    experience_levels = skills_pivot['experience'].unique()
    for experience_level in experience_levels:
        subset = skills_pivot[skills_pivot['experience'] == experience_level].nlargest(10, 'id')

        plt.figure(figsize=(6, 4))
        plt.barh(subset['key_skills'], subset['id'], color='skyblue')
        plt.xlabel('Количество упоминаний')
        plt.ylabel('Навыки')
        plt.title(f'{title} - {experience_level}')
        plt.gca().invert_yaxis()
        plt.show()

# Визуализация топ-10 навыков для Data Analyst по уровню опыта
visualize_top_skills_by_experience(key_skills_da_pivot, 'Топ-10 навыков для Аналитика данных')

# Визуализация топ-10 навыков для Data Scientist по уровню опыта
visualize_top_skills_by_experience(key_skills_ds_pivot, 'Топ-10 навыков для Data Scientist')
```

### Расчет помесячной динамики количества вакансий для Аналитика данных и Бизнес-аналитика. Ответ отдельно дайте для грейдов Junior, Junior+, Middle, Senior.

```python
vacancies_da_cleaned['published_at'] = pd.to_datetime(vacancies_da_cleaned['published_at'])
vacancies_ds_cleaned['published_at'] = pd.to_datetime(vacancies_ds_cleaned['published_at'])

# Добавление столбца с месяцем и годом
vacancies_da_cleaned['year_month'] = vacancies_da_cleaned['published_at'].dt.to_period('M')
vacancies_ds_cleaned['year_month'] = vacancies_ds_cleaned['published_at'].dt.to_period('M')

# Функция для расчета помесячной динамики количества вакансий
def calculate_monthly_dynamics(data, title):
    dynamics = data.groupby(['year_month', 'experience']).size().unstack(fill_value=0)
    dynamics.plot(kind='line', figsize=(8, 4), title=f'Помесячная динамика количества вакансий для {title}')
    plt.xlabel('Месяц')
    plt.ylabel('Количество вакансий')
    plt.show()

# Расчет помесячной динамики для Data Analysis
calculate_monthly_dynamics(vacancies_da_cleaned, 'Data Analysis')

# Расчет помесячной динамики для Data Science
calculate_monthly_dynamics(vacancies_ds_cleaned, 'Data Science')
```
